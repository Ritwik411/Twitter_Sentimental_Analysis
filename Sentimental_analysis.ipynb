```
pos_tweet_tok = twitter_samples.tokenized('positive_tweets.json')
neg_tweet_tok = twitter_samples.tokenized('negative_tweets.json')
#neutral_tok = twitter_samples.tokenized('tweets.20150430-223406.json')
#neutral_tok = neutral_tok[:5000]
```

```
from nltk.tag import pos_tag
from nltk.stem.wordnet import WordNetLemmatizer
```

```
def clean(tweet_tokens):
  import re,string
  clean_tokens = []
  stop_words = stopwords.words('english')
  for token,tag in pos_tag(tweet_tokens):
        token = re.sub('(RT)|http[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+#]|[!*\(\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))+','', token)
        token = re.sub("(@[A-Za-z0-9_]+)","", token)
        if tag.startswith("NN"):
            pos = 'n'
        elif tag.startswith('VB'):
            pos = 'v'
        else:
            pos = 'a'

        lemmatizer = WordNetLemmatizer()
        token = lemmatizer.lemmatize(token, pos)

        if len(token) > 0 and token not in string.punctuation and token.lower() not in stop_words:
            clean_tokens.append(token.lower())
  return clean_tokens
```

```
p_clean = []
n_clean = []
#neutral_clean = []
for i in pos_tweet_tok:
  p_clean.append(clean(i))
for i in neg_tweet_tok:
  n_clean.append(clean(i))
# for i in neutral_tok:
#   neutral_clean.append(clean(i))
```

```
pos_words = [word for i in p_clean for word in i]
neg_words = [word for i in n_clean for word in i]
# neutral_words = [word for i in neutral_clean for word in i]
```
